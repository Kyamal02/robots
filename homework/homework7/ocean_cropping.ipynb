```python
import cv2
import matplotlib.pyplot as plt

pic = cv2.imread('ocean.jpg')
resizeScale = 0.9

# resize and show
img = cv2.resize(pic, None, fx=resizeScale, fy=resizeScale, interpolation=cv2.INTER_LINEAR)
cv2.imshow('', img)
cv2.waitKey(20)
```

```python
# height / 3, width / 2, number of channels
height, width, channelsN = img.shape
print(height, width, channelsN)

h0 = w0 = 1
picCounter = 1
h_interval = height // 3
w_interval = width // 3

# let's cropp
for h in range(h_interval-1, height-1, h_interval):
    w0 = 1
    for w in range(w_interval-1, width-1, w_interval):
        cropped_pic = img[h0:h, w0:w]
        cv2.imwrite(f"cropped_ocean/Cropped Image{picCounter}.jpg", cropped_pic)
        w0 = w
        picCounter += 1
    h0 = h
```

```python
cv2.destroyAllWindows()
size = 2
# show cropped pictures
f, axarr = plt.subplots(size, size)
for i in range(1, picCounter):
    pic = cv2.imread(f'cropped_ocean/Cropped Image{i}.jpg')
    axarr[i // size - 1, i % size - 1].imshow(cv2.cvtColor(pic, cv2.COLOR_RGB2HLS))
    axarr[i // size - 1, i % size - 1].axis('off')
plt.show()
```
